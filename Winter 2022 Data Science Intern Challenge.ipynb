a. Think about what could be going wrong with our calculation. Think about a better way to evaluate this data.

lets see How is 3145.13 calculated?

```python
df['order_amount'].describe()
```

```python
#Order Amount Sum
order_amount_sum = df['order_amount'].sum()

#Sum of Items
total_items_count = df['total_items'].count()
 
AOV = order_amount_sum / total_items_count
print('$', str(round(AOV, 2)))
```

This calculation of ours could found incorrect because we are dividing by the total number of rows(5000). Count() is returning the no. of rows in column total_items, Not the actual no. of items(4336). The actual average order value will be calculated by dividing the sum of items Amount( Sum of order_amount) by the total no. of items (Sum of all total_items). lets calculate it in that way.

```python
#Order Amount Sum
order_amount_sum = df['order_amount'].sum()

#Total no of Items
total_items_sum = df['total_items'].sum()

actual_AOV = order_amount_sum / total_items_sum
print('$', str(round(actual_AOV, 2)))
```

There are 17 Such Orders with total_items 2000 and order_amount $704000. These are affecting the average order_amount we are concerned about.A better way to evaluate the data is by removing all the purchases made by this customers and re-evaluating the AOV of the data. When outliers have been removed the calculated AOV becomes approximately 302.6 dollars.
